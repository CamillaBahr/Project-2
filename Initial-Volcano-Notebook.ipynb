```python
# Dependencies
import requests
import json
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
import socket
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine
```

```python
url ="https://webservices.volcano.si.edu/geoserver/GVP-VOTW/wms?service=WFS&version=1.0.0&request=GetFeature&typeName=GVP-VOTW:E3WebApp_Eruptions1960&outputFormat=application%2Fjson"
```

```python
response = requests.get(url).json()
print(json.dumps(response, indent=4, sort_keys=True))
```

```python
features = []
lat = []
long = []
indexes = []
indexes = list(range(0,2074))
name = []
active = []
explosive = []
startyear = []
startmonth = []



response = requests.get(url).json()
for index in indexes:
    lat.append(response['features'][index]['properties']['LatitudeDecimal'])
    long.append(response['features'][index]['properties']['LongitudeDecimal'])
    name.append(response['features'][index]['properties']['VolcanoName'])
    active.append(response['features'][index]['properties']['ContinuingEruption'])
    explosive.append(response['features'][index]['properties']['ExplosivityIndexMax'])
    startyear.append(response['features'][index]['properties']['StartDateYear'])
    startmonth.append(response['features'][index]['properties']['StartDateMonth'])
    
```

```python
volcano_dict = {
    "name": name,
    "lat": lat,
    "long": long,
    "active": active,
    "explosive": explosive,
    "startyear": startyear,
    "startmonth": startmonth
}
volcano_data = pd.DataFrame(volcano_dict)
volcano_data
```

```python
volcano_data = volcano_data.sort_values('startyear', ascending=False)
volcano_data
```

```python
volcano_data = volcano_data[volcano_data['startyear'] > 2010]
volcano_data
```

```python
volcano_data = volcano_data.rename(columns={'lat': 'latitude', 'long': 'longitude'})
volcano_data
```
